```python
import pandas as pd
```

```python
train = pd.read_csv("C:\\Users\LENOVO\\Desktop\\quera ex5\\diamond\\data\\diamonds_train.csv")
train
```

```python
test = pd.read_csv("C:\\Users\\LENOVO\\Desktop\\quera ex5\\diamond\\data\\diamonds_test.csv")
test
```

```python
train.value_counts()
```

```python
train.isna().sum()
```

```python

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
train['cut'] = label_encoder.fit_transform(train['cut'])
train['color'] = label_encoder.fit_transform(train['color'])
train['clarity'] = label_encoder.fit_transform(train['clarity'])
train
```

```python
train.head()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
x_train = train.drop(columns=['price'])
y_train = train['price']
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score
y_pred = model.predict(x_train)
r2_score(y_train, y_pred)
```

```python
label_encoder = LabelEncoder()
test['cut'] = label_encoder.fit_transform(test['cut'])
test['color'] = label_encoder.fit_transform(test['color'])
test['clarity'] = label_encoder.fit_transform(test['clarity'])
test
```

```python
x_test = test
y_test_pred = model.predict(x_test)
```

```python
submission = pd.DataFrame({'price': y_test_pred})
```

```python
import zipfile
import joblib

def compress(file_names):
    print("File Paths:")
    print(file_names)
    compression = zipfile.ZIP_DEFLATED
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            zf.write('./' + file_name, file_name, compress_type=compression)

submission.to_csv('submission.csv', index=False)
joblib.dump(model, 'model')
file_names = ['diamond.ipynb', 'submission.csv', 'model']
compress(file_names)
```
