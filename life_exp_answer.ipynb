```python
test = pd.read_csv("C:\\Users\\LENOVO\\Desktop\\quera ex6\\data\\test.csv")
test.info()
```

```python
hep_mean = train['Hepatitis B'].mean()
test['Hepatitis B'] = test['Hepatitis B'].fillna(value=hep_mean)
tot_mean = train['Total expenditure'].mean()
test['Total expenditure'] = test['Total expenditure'].fillna(value=tot_mean)
alc_mean = train['Alcohol'].mean()
test['Alcohol'] = test['Alcohol'].fillna(value=alc_mean)
gdp_mean = train['GDP'].mean()
test['GDP'] = test['GDP'].fillna(value=gdp_mean)
test.info()
```

```python
# ...existing code...
from sklearn.preprocessing import LabelEncoder
test = test.drop(columns=['Schooling', 'infant deaths', 'Year', 'Country'])
label_encoder = LabelEncoder()
test['Status'] = label_encoder.fit_transform(test['Status'])
```

```python
# ...existing code...
train = train.drop(columns=['Schooling', 'infant deaths', 'Year'])
train['GDP'] = train['GDP'].fillna(train.groupby('Country')['GDP'].transform('mean'))
train['GDP'] = train['GDP'].fillna(train['GDP'].mean())
train['Hepatitis B'] = train['Hepatitis B'].fillna(train.groupby('Country')['Hepatitis B'].transform('mean'))
train['Hepatitis B']  = train['Hepatitis B'].fillna(train['Hepatitis B'].mean())
train['Total expenditure'] = train['Total expenditure'].fillna(train.groupby('Country')['Total expenditure'].transform('mean'))
train['Total expenditure'] = train['Total expenditure'].fillna(train['Total expenditure'].mean())
train['Alcohol'] = train['Alcohol'].fillna(train.groupby('Country')['Alcohol'].transform('mean'))
train['Alcohol'] = train['Alcohol'].fillna(train['Alcohol'].mean())
train['BMI'] = train['BMI'].fillna(train.groupby('Country')['BMI'].transform('mean'))
train['BMI'] = train['BMI'].fillna(train['BMI'].mean())
train['thinness  1-19 years'] = train['thinness  1-19 years'].fillna(train.groupby('Country')['thinness  1-19 years'].transform('mean'))
train['thinness  1-19 years'] = train['thinness  1-19 years'].fillna(train['thinness  1-19 years'].mean())
train['Polio'] = train['Polio'].fillna(train.groupby('Country')['Polio'].transform('mean'))
train['Polio'] = train['Polio'].fillna(train['Polio'].mean())
train['Diphtheria'] = train['Diphtheria'].fillna(train.groupby('Country')['Diphtheria'].transform('mean'))
train['Diphtheria'] = train['Diphtheria'].fillna(train['Diphtheria'].mean())
train['Population'] = train['Population'].fillna(train.groupby('Country')['Population'].transform('mean'))
train['Population'] = train['Population'].fillna(train['Population'].mean())

# ...existing code...
train.info()
```

```python
train = train.drop(columns=['Country'])
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
train['Status'] = label_encoder.fit_transform(train['Status'])

train
```

```python
from sklearn.preprocessing import PolynomialFeatures
x_train = train.drop(columns=['Life expectancy'])
x_test = test
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
x_train = std_scaler.fit_transform(x_train)
x_test = std_scaler.transform(test)

poly_transformer = PolynomialFeatures(degree=2, include_bias=True)
x_train_trans = poly_transformer.fit_transform(x_train)
x_test_trans = poly_transformer.transform(x_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
x_train = train.drop(columns=['Life expectancy'])
y_train = train['Life expectancy']
poly_transformer = PolynomialFeatures(degree=2 , include_bias=True)
x_trains_trans = poly_transformer.fit_transform(x_train)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score

y_pred = model.predict(x_train)
r2_score(y_train, y_pred)
```

```python
x_test = test
y_test_pred = model.predict(x_test)
```

```python
submission = pd.DataFrame({'Life expectancy' : y_test_pred})
submission
```

```python
import zipfile
import joblib

def compress(file_names):
    print("File Paths:")
    print(file_names)
    compression = zipfile.ZIP_DEFLATED
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            zf.write('./' + file_name, file_name, compress_type=compression)


joblib.dump(poly_transformer, 'poly_transformer')
joblib.dump(model, 'model')
submission.to_csv('submission.csv', index=False)

file_names = ['poly_transformer', 'model', 'submission.csv', 'life_expectancy.ipynb']
compress(file_names)
```
